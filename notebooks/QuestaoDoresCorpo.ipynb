```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Carregar os dados relacionados à questão de qual região do corpo sente dor

```python
data = pd.read_excel(io = 'EDA_V1.xlsx', sheet_name = 'QuestaoDoCorpo')
```

### Transformas os  dados de linhas para colunas

```python
meltdata = pd.melt(data)
```

### Eliminar as linhas com dados faltantes

```python
meltdata = meltdata.dropna()
```

### Construir Histrograma para o número de vezes respondido cada parte do corpo para dor

```python
sns.distplot(meltdata['value'])
```

### As dores mais presentes foram:
- "Lombar central" 
- "Lombar esquerda"
- "Panturrilha da perna esquerda"
- "Região traseira do joelho esquerda"
